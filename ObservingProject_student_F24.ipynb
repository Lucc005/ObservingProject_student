<p>This requires a bit more work because you want to look at the color-magnitude diagram of the cluster to get some sense of how faint you need to observe to get a decent chunk of the CMD. For this calculation, assume the gain is 1.  This means that if I get one photon, my CCD counts 1 electron.  Since Poisson noise is computed on the incident photons and not the number of measured electrons, you need to convert your counts to photons.  We will assume for now that this is a 1-to-1 conversion (Gain=1).  This is not correct but will be ok for this calculation.

```python
# Answer 7) Here
# YOUR CODE GOES HERE

# Hint: You need to use the information about the number of counts 
# for different sources that I give above.
# In each band, think about the noise from the background over the 
# whole size of the source when observed in 15 minutes. From this and the 
# SNR requirement, you can determine the number of object counts that you need 
# to have the required SNR (see lecture notes). Finally, the flux from a 
# 20th mag object can then be used to convert that number of counts to a 
# limiting magnitude.

# Define constants for exposure time (15 minutes), pixels, and read noise
exposure_time_seconds = 900  # Exposure time of 15 minutes, converted to seconds (900 seconds)
pixels = 10  # Number of pixels over which the object counts are summed. The object spans 10 pixels
read_noise = 3.7  # The intrinsic read noise (in counts) for each pixel. It is constant for the detector.

# Background counts for the B and V bands during a 100-second exposure
background_B_100s = 194.2  # The background counts in the B-band (Blue band) per pixel for a 100-second exposure
background_V_100s = 144.8  # The background counts in the V-band (Visual band) per pixel for a 100-second exposure

# Convert background counts for 100 seconds to the 15-minute exposure time (900 seconds)
# We need to scale these background counts to match the 900-second exposure time.
# The total counts for 15 minutes are calculated by scaling the counts per 100 seconds
# by the ratio of exposure times (900 seconds / 100 seconds), and then multiplying by 
# the number of pixels over which the object is summed.
background_B_15min = (background_B_100s / 100) * exposure_time_seconds * pixels  # Background counts in B-band for 15 minutes
background_V_15min = (background_V_100s / 100) * exposure_time_seconds * pixels  # Background counts in V-band for 15 minutes

# Function to calculate the noise from the background
# The total noise comes from two main sources:
# 1. Background noise, which is directly proportional to the number of background counts.
# 2. Read noise, which is constant for each pixel and depends on the number of pixels summed.
def calculate_noise(background_counts):
    # The total noise is the square root of the sum of the background noise and the read noise squared.
    # The read noise is multiplied by the number of summed pixels to account for the contribution
    # of read noise across the full area of the object.
    return np.sqrt(background_counts + (read_noise ** 2) * pixels)

# Calculate the background noise in both the B and V bands
noise_B = calculate_noise(background_B_15min)  # Noise from background in the B-band
noise_V = calculate_noise(background_V_15min)  # Noise from background in the V-band

# Set the desired signal-to-noise ratio (SNR) for detection
# SNR is a measure of how detectable an object is relative to the background noise.
# A higher SNR means that the object is more clearly distinguishable from the background.
desired_SNR = 20  # We want the signal from the object to be 20 times larger than the noise.

# Calculate the required number of object counts to achieve the desired SNR
# To determine how many object counts are needed to achieve the desired SNR,
# we use the formula derived from the SNR requirement:
# counts = 0.5 * (SNR^2 + sqrt(SNR^4 + 4 * SNR^2 * background_noise))
counts_B_SNR = 0.5 * (desired_SNR**2 + np.sqrt(desired_SNR**4 + 4 * desired_SNR**2 * background_B_15min))  # Required object counts for B-band
counts_V_SNR = 0.5 * (desired_SNR**2 + np.sqrt(desired_SNR**4 + 4 * desired_SNR**2 * background_V_15min))  # Required object counts for V-band

# Calculate the counts for a 20th magnitude star in 15 minutes
# The counts for a 20th magnitude star are scaled from a 100-second exposure to the 15-minute exposure time.
# This scaling accounts for the fact that the same star will contribute more counts in a longer exposure time.
star_counts_B_15min = (889 / 100) * exposure_time_seconds  # Counts for a 20th magnitude star in the B-band for 15 minutes
star_counts_V_15min = (953 / 100) * exposure_time_seconds  # Counts for a 20th magnitude star in the V-band for 15 minutes

# Function to calculate the magnitude corresponding to a given number of counts
# The formula for converting counts to magnitude is based on the logarithmic relationship between flux and magnitude:
# magnitude = -2.5 * log10(counts / reference_counts) + reference_magnitude
# Here, reference_counts corresponds to the counts from a 20th magnitude star in the respective band.
def magnitude_from_counts(C, B):
    # Return the magnitude corresponding to the given number of counts.
    return -2.5 * np.log10(C / B) + 20  # Calculate magnitude using the formula based on counts

# Calculate the limiting magnitudes corresponding to the SNR limits in both the B and V bands
# The limiting magnitude is the faintest magnitude that can be detected with the desired SNR.
# It is calculated by comparing the required object counts for a given SNR with the counts from a 20th magnitude star.
limiting_magnitude_B = magnitude_from_counts(counts_B_SNR, star_counts_B_15min)  # Limiting magnitude in the B-band
limiting_magnitude_V = magnitude_from_counts(counts_V_SNR, star_counts_V_15min)  # Limiting magnitude in the V-band

# Print the results to show the values calculated at each step:

# a) The total background counts in both the B and V bands over 15 minutes
# This shows how much background light contributes to the noise in the observation.
print(f"a) The number of BKG counts in B and V in 15min is: {background_B_15min:4.2f}, {background_V_15min:4.2f}")

# b) The total background noise in both the B and V bands
# This shows the combined effect of the background light and the detector's intrinsic read noise.
print(f"b) BKG noise in B and V: {noise_B:4.2f}, {noise_V:4.2f}")

# c) The limiting object counts in both the B and V bands needed to achieve the desired SNR
# These are the minimum number of object counts required for the object to be detectable with the desired SNR.
print(f"c) limiting counts for SNR limit in B and V: {counts_B_SNR:4.2f}, {counts_V_SNR:4.2f}")

# d) The number of counts for a 20th magnitude star in both the B and V bands over 15 minutes
# This shows how many counts a 20th magnitude star would contribute in the given exposure time.
print(f"d) the number of counts in B and V for a 20th mag object in 15min: {star_counts_B_15min:5.2f}, {star_counts_V_15min:5.2f}")

# e) The limiting magnitudes corresponding to the SNR limit in both the B and V bands
# These are the faintest magnitudes that can be detected at the given SNR threshold.
print(f"e) magnitude limits in B and V corresponding to our SNR limit: {limiting_magnitude_B:4.2f}, {limiting_magnitude_V:4.2f}")
```

```python
#***RUN THIS BLOCK. If you wish to change the font sizes in the plots, you can do so here.  ***

#Now we just define some settings for our plots. 
params = {
    #'text.latex.preamble': ['\\usepackage{gensymb}'],
    'image.origin': 'lower',
    'image.interpolation': 'nearest',
    'image.cmap': 'gray', #gray
    'axes.grid': False,
    'savefig.dpi': 300,  
    'axes.labelsize': 14, # fontsize for x and y labels
    'axes.titlesize': 20, 
    'font.size': 14, 
    'legend.fontsize': 10, 
    'xtick.labelsize': 12,
    'ytick.labelsize': 12, 
    'text.usetex': False,
    'figure.figsize': [6, 6], #[width, height]
    'figure.autolayout': False,
    'font.family': 'monospace',
}
matplotlib.rcParams.update(params) #activate the settings

def make_plots(file_name):
    print("********************************************")
    print("Plots for file: "+file_name)
    print("********************************************")
    #We load in the first 4 columns of the file
    data = np.loadtxt(file_name,skiprows = 0,usecols = (0,1,2,3), dtype={'names': ('No', 'Ref', 'V', 'B-V'), 'formats': ('S8', 'S16', 'S16','S16')})
    
    #We need to handle duplicates in the file, so we find indeces of unique sources
    indeces_use = []
    len_data = len(data)
    for i in range(1,len_data-1): #Skip first row since it contains columns names.
        if data[i][0] != data[i+1][0]: #Check if the next source is a repeat
            indeces_use.append(i) #If it is not a repeat, it is unique and we record its index.
            
    num_sources = len(indeces_use) #Number of actual unique sources
    
    #We will now average over the magnitudes for repeated sources
    V_avg = np.zeros(num_sources)
    BV_avg = np.zeros(num_sources)
    for i in range(num_sources):
        current_index = indeces_use[i] #Keep track of the first index that this source appears at
        num_i = 1 #Keep track of the number of sources with this source name
        V_tot = float(data[current_index][2])
        BV_tot = float(data[current_index][3])
        #While the next source has the same source name
        while current_index < (num_sources - 2) and data[current_index][0] == data[current_index+1][0]:
            current_index += 1 #We go to the next repeat of this source.
            num_i += 1 
            V_tot += float(data[current_index][2]) #Add the next value forn this repeated source to the total mag
            BV_tot += float(data[current_index][3])
        V_avg[i] = V_tot/num_i
        BV_avg[i] = BV_tot/num_i

    #Make the figure for the CMD
    fig1, ax1 = plt.subplots()
    ax1.set_xlabel(r'$B-V\ [mag]$')
    ax1.set_ylabel(r'$V\ [mag]$')
    ax1.grid(which= 'major', linestyle='-', linewidth='0.5', color = 'black', alpha = 0.3)
    ax1.grid(which='minor', linestyle=':', linewidth='0.3', color = 'black', alpha = 0.2)
    ax1.scatter(BV_avg, V_avg, marker = '.', c='k')

    #Create good min and max values for CMD axes, and set axis limits. 
    x_min_plot = np.min(BV_avg)-abs(.1*np.min(BV_avg))
    x_max_plot = 1.1*np.max(BV_avg)
    y_min_plot = np.min(V_avg)-abs(.1*np.min(V_avg))
    y_max_plot = 1.1*np.max(V_avg)
    ax1.set_xlim([float(x_min_plot), float(x_max_plot)])
    ax1.set_ylim([float(y_max_plot), float(y_min_plot)])
    ax1.set_title(r'Color Magnitude Diagram')

    #Make figure for histogram. 
    fig2, ax2 = plt.subplots()
    ax2.hist(V_avg, bins = 'auto',histtype ='step', color = 'k', lw = 1.5)
    ymin_hist, ymax_hist = ax2.get_ylim()
    xmin_hist, xmax_hist = ax2.get_xlim()
    #Write on the number of total sources for reference
    ax2.annotate( 'N = '+str(num_sources), xy = ( xmin_hist + 0.05*(xmax_hist-xmin_hist), ymax_hist - 0.1*ymax_hist))
    ax2.set_title(r'V mag Histogram')
    ax2.set_xlabel(r'$V\ [mag]$')
    ax2.set_ylabel(r'Counts')
```
